```python
import numpy as np
import matplotlib.pyplot as pl
import pandas as pd
from astropy.table import Table
%matplotlib inline
```

```python
mask = [False, False, True, True, False, False, False, True, True]
header = ["epic_number", "mix_mean", "mix_sd", "mix_neff", "mix_rhat", "logdeltaQ_mean", "logdeltaQ_sd", "logdeltaQ_neff", "logdeltaQ_rhat",
         "logQ0_mean", "logQ0_sd", "logQ0_neff", "logQ0_rhat", "logperiod_mean", "logperiod_sd", "logperiod_neff", "logperiod_rhat",
         "logamp_mean", "logamp_sd", "logamp_neff", "logamp_rhat", "logs2_mean", "logs2_sd", "logs2_neff", "logs2_rhat", "acfpeak"]
with open("../output/out.dat", "r") as file:
    lines = file.readlines()
    nstars = (np.int((len(lines)/7)))
    data = np.zeros((nstars, 26))
    for i in range(nstars):
        data[i, 0] = lines[7*i+1].split()[0]
        for j in range(6):
            data[i, 4*j+1:4*(j+1)+1] = np.array(lines[7*i+j+1].split())[mask]
        acfpeak = lines[7*i+7].split()[2]
        if "None" in acfpeak:
            data[i, 25] = np.nan
        else:
            data[i, 25] = acfpeak
```

```python
pd.DataFrame(data=data, columns=header)
```

```python
df = pd.read_table("../output/out.dat", sep="\t", header=0, dtype={'EPIC number': int, 'variable name': str, 'mean':str, 'sd':float, 'mc_error':float, 'hpd_2.5':float, 'hpd_97.5':float, 'n_eff':str, 'Rhat':float})
df = df.rename(columns = lambda x: x.strip())
df["n_eff"] = np.array(df["n_eff"], dtype=float)
df.loc[df["mean"].str.contains("None"), "mean"] = "NaN"
df["mean"] = np.array(df["mean"], dtype=float)
```

```python
has_acfpeak = np.isfinite(df.loc[df["variable name"].str.contains("acfpeak")]["mean"])
gp_acf_correspond = np.isclose(df.loc[df["variable name"].str.contains("acfpeak")]["mean"], df.loc[df["variable name"].str.contains("logperiod")]["mean"], rtol=1)
converged = np.array(df.loc[df["variable name"].str.contains("logperiod")]["Rhat"] < 1.01)
```
